```python
base_dados = pd.read_csv('cancelamentos.csv').drop(columns='CustomerID').dropna()
display(base_dados['cancelou'].value_counts(normalize=True))

base_dados = base_dados[base_dados['duracao_contrato'] != 'Monthly']
display(base_dados['cancelou'].value_counts(normalize=True))

base_dados = base_dados[base_dados['ligacoes_callcenter'] <=4]
display(base_dados['cancelou'].value_counts(normalize=True))

base_dados = base_dados[base_dados['dias_atraso'] <= 20]
display(base_dados['cancelou'].value_counts(normalize=True))
```
