```python
import pandas as pd
import gdown
import tempfile
import os
import uuid

#Baixando o dataset do Google Drive e lendo-o em um DataFrame
temp_dir = tempfile.gettempdir()
temp_path = os.path.join(temp_dir, f"dataset_{uuid.uuid4().hex}.parquet")
try:
    gdown.download(id='13KRdiYnQikDx3pxPKGUTeejOz1fLHjzm', output=temp_path, quiet=True)
    df = pd.read_parquet(temp_path)
finally:
    if os.path.exists(temp_path):
        os.remove(temp_path)

display(df.head())
display(df.info())
```
